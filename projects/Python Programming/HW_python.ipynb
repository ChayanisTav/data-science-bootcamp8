```python
## HW01 pao ying chub game
import random

def play_pyc():
    user_score = 0
    computer_score = 0

    print("Hello, player!")
    name = input("What's your name? ")
    print(f"Hi {name}, my name is Computer.")

    ready = input("Are you ready (yes/no)? ").lower()

    if ready == "yes":
        print("This is a PAO YING CHUB game.")
        print("You need to guess -> rock, paper, or scissors to win the game.")
        print("If you win, you will get 1 point.")
        print("If you lose, you will lose 1 point.")
        print("If you want to exit the game, type 'exit'")
        print("Let's start!")

        while True:
            print("Your turn")
            user_choice = input("Choose rock, paper, or scissors: ").lower()

            choices = ["rock", "paper", "scissors"]

            # Check if user wants to quit
            if user_choice == "quit":
                break

            # Validate user input if not quitting
            if user_choice not in choices:
                print("Invalid choice. Please choose from rock, paper, or scissors.")
                continue

            # Continue the game if user doesn't want to quit and input is valid
            computer_choice = random.choice(choices)

            if computer_choice == "rock":
                outcome = "It's a tie!" if user_choice == "rock" else "You win!" if user_choice == "paper" else "You lose!"
            elif computer_choice == "paper":
                outcome = "You lose!" if user_choice == "rock" else "It's a tie!" if user_choice == "paper" else "You win!"
            else:
                outcome = "You win!" if user_choice == "rock" else "You lose!" if user_choice == "paper" else "It's a tie!"

            # Update scores based on the outcome
            if outcome == "You win!":
                user_score += 1
            elif outcome == "You lose!":
                computer_score += 1

            # Display individual outcomes
            print(f"\nComputer's choice: {computer_choice}")
            print(outcome)
            print(f"Your score: {user_score}")
            print(f"Computer score: {computer_score} \n")

        # Display final scores and determine the winner
        print("\nThanks for playing my game")
        print("\nFinal Scores\n")
        print(f"You: {user_score}")
        print(f"Computer: {computer_score}")

        if user_score > computer_score:
            print("Congratulations! You won the game!")
        elif user_score < computer_score:
            print("Sorry, you lost the game.")
        else:
            print("It's a tie!")

    else:
        print("Thanks for coming.")

play_pyc()
```

```python
## HW02 class ATM at least 5
class ATM:
    def __init__(self, balance=0):
        self.balance = balance

    def view_money(self):
        return f"Current balance: ${self.balance}"

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"Withdrawal successful. Remaining balance: ${self.balance}"
        elif amount <= 0:
            return "Invalid withdrawal amount. Please enter a positive value."
        else:
            return "Insufficient funds. Withdrawal failed."

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposit successful. New balance: ${self.balance}"
        else:
            return "Invalid deposit amount. Please enter a positive value."

    def option_select(self, option):
        if option == 1:
            return self.view_money()
        elif option == 2:
            amount = float(input("Enter withdrawal amount: $"))
            return self.withdraw(amount)
        elif option == 3:
            amount = float(input("Enter deposit amount: $"))
            return self.deposit(amount)
        else:
            return "Invalid option. Please select a valid option."

    def get_balance(self):
        return self.balance

atm = ATM(0)

while True:
    print("\nOptions:")
    print("1. View Money")
    print("2. Withdraw")
    print("3. Deposit")
    print("4. Exit")
    choice = int(input("Select an option (1-4): "))

    if choice == 4:
        break

    result = atm.option_select(choice)
    print(result)

# Display final balance
print(f"\nFinal Balance: ${atm.get_balance()}")
```

```python
## HW03 imdb and clean data web scraping
!pip install gazpacho
from gazpacho import Soup
import requests

url = "https://editorial.rottentomatoes.com/guide/popular-movies/"
html = requests.get(url)

rotten = Soup(html.text)
titles = rotten.find("div", {"class":"article_movie_title"})
clean_titles = [title.strip() for title in titles]
years = rotten.find("span", {"class":"subtle start-year"})
clean_years = [year.strip() for year in years]

import pandas as pd
## create dataframe
rotten_tomato_movie_database = pd.DataFrame(data = {
    "title": clean_titles,
    "years" : clean_years
})
rotten_tomato_movie_database.head()
```

```python
## HWO4 find public api use python to get data
import requests

# Replace with your OpenWeatherMap API key
api_key = "YOUR_API_KEY"

# Set the city and parameters
city = "Bangkok"
params = {
    "q": city,
    "appid": api_key
}

# Send the request to the OpenWeatherMap API
response = requests.get("https://api.openweathermap.org/data/2.5/weather", params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()

    # Extract the weather data
    weather = data['weather'][0]
    description = weather['description']
    temperature = data['main']['temp']

    # Print the weather data
    print(f"The weather in {city} is currently {description} with a temperature of {temperature} degrees Celsius.")
else:
    print("Failed to fetch weather data.")
```
